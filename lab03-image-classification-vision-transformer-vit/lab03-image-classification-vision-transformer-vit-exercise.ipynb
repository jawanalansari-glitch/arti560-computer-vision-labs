```python
# Import Libraries
import numpy as np
import keras
import keras_hub

# Load ViTImageClassifierPreprocessor (required by the lab)
preprocessor = keras_hub.models.ViTImageClassifierPreprocessor.from_preset(
    "vit_base_patch16_224_imagenet"
)

# Load ViTImageClassifier (vit_base_patch16_224_imagenet preset)
model = keras_hub.models.ViTImageClassifier.from_preset(
    "vit_base_patch16_224_imagenet"
)

# Load ImageNet class names
with open("imagenet_classes.txt", "r") as f:
    class_names = [line.strip() for line in f.readlines()]

# Load the images
image_files = ["image1.jpg", "image2.jpg", "image3.jpg", "image4.jpg", "image5.jpg"]
true_labels = ["banana", "airplane", "bicycle", "soccer ball", "cat"]

images = []
for file in image_files:
    img = keras.utils.load_img(file, target_size=(224, 224))
    img = keras.utils.img_to_array(img)
    images.append(img)

images = np.array(images, dtype="float32")

# Predict classes
# Do NOT call preprocessor(images) here, because the preset classifier already has preprocessing built in
predictions = model.predict(images)

predicted_indices = np.argmax(predictions, axis=1)
predicted_labels = [class_names[i] for i in predicted_indices]

# Print predictions
for file, pred, true in zip(image_files, predicted_labels, true_labels):
    print(f"Image File: {file}")
    print(f"Predicted Label: {pred}")
    print(f"True Label: {true}")
    print("-" * 40)
```
